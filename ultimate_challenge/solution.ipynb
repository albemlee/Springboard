# Part 2 ‑ Experiment and metrics design

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.
However, a toll bridge, with a two­way toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have
proposed an experiment to encourage driver partners to be available in both cities, by
reimbursing all toll costs.

#### What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric?

I would track the proportion of drivers who are have been active in both cities on at least 10 days within the last month. Driver activity may vary from day to day, and drivers active in both cities may not be active in both cities on all days.

#### Describe a practical experiment you would design to compare the effectiveness of the proposed change in relation to the key measure of success. a. how you will implement the experiment b. what statistical test(s) you will conduct to verify the significance of the observation c. how you would interpret the results and provide recommendations to the city operations team along with any caveats.

A. How you will implement the experiment?

I would first prepare my data and ensure I have a labeled describing whether driver partners are exclusive to one city. I would then randomly select half of the driver partners and provide them with  a voucher to reimburse toll costs for a month. Each day I would label driver partners on whether they were active in both cities, and I expect that many driver partners will stop using the voucher after trying it the first few times. At the end of the month, I would label driver partners on whether they were active in both cities on at least 10 days.

B. What statistical test(s) you will conduct to verify the significance of the observation?

Assuming that I have a large sample of driver partner activity, I would perform a test comparing the activity of driver partners without vouchers to the activity of driver partners with vouchers.

My hypotheses would be:
- H_0: The proportion of driver partners active in both cities is similar between those with and without vouchers.
- H_a: The proportion of driver partners active in both cities is higher for those with vouchers than those without.

C. How you would interpret the results and provide recommendations to the city operations team along with any caveats.

If the p-value ends up being less than 0.05, I will reject the null hypothesis and conclude that The proportion of driver partners active in both cities is higher for those with vouchers than those without.

If the vouchers were indeed effective, I would then tell city operations that the vouchers may be an effective way to increase the driver partners active in both cities. I would make sure to emphasize that vouchers were used during the experiment, and changing the method of reducing toll costs will change the user experience which may lead to unexpected results. Depending on their interest in exploring other methods of reducing toll costs, I may recommend addition experiments that test those methods.

While city operations is implementing the change, I will make sure that they track the proportion of driver partners who have been active in both cities on at least 10 days in the past month.

## Part 2

Build a predictive model to help Ultimate determine whether or not a user will be active
in their 6th month on the system. Discuss why you chose your approach, what
alternatives you considered, and any concerns you have. How valid is your model?
Include any key indicators of model performance

```python
ultimate_data_challenge['last_trip_date'].max()
```

```python
# Filter to users who have been signed up for at least 6 months
old_users = ultimate_data_challenge[
    ultimate_data_challenge['signup_date'] <= ultimate_data_challenge['last_trip_date'].max() - pd.DateOffset(months=6)
].copy()
# Filter to users who have been signed up for less than 6 months
new_users = ultimate_data_challenge[
    ultimate_data_challenge['signup_date'] > ultimate_data_challenge['last_trip_date'].max() - pd.DateOffset(months=6)
].copy()
print(old_users.shape)
print(new_users.shape)
```

#### Create Label

```python
def calc_active_user(signup_date, last_trip_date):
    # 6th month is after 5 full months
    if last_trip_date >= signup_date + pd.DateOffset(months=5):
        return True
    else:
        return False
    
old_users['active_user'] = [
    calc_active_user(signup_date, last_trip_date) 
    for signup_date, last_trip_date in zip(old_users['signup_date'], old_users['last_trip_date'])
]
```

```python
old_users['active_user'].sum()
```

Only 706 out of 1737 riders are still active in month 6

#### Feature Selection

correlations within features

```python
#One Hot Encode Categorical Features
old_users_clf = pd.get_dummies(old_users, dummy_na=True)

features = [
    'avg_dist', 
    'avg_rating_by_driver', 
    'avg_rating_of_driver', 
    'avg_surge', 
    'surge_pct', 
    'trips_in_first_30_days', 
    'weekday_pct',
    'days_signup', 
    'city_Astapor', 
    "city_King's Landing", 
    'city_Winterfell',
    'phone_Android', 
    'phone_iPhone',
    'phone_nan'
]

X = old_users_clf[features]
y = old_users_clf['active_user']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size= 0.2)
```

```python
clf = XGBClassifier()
clf.fit(X, y)
```

```python
from xgboost import plot_importance
plot_importance(clf)
```

Phone, city, and surge_pct did not seem to be strong predictors, so they will be removed from the model.

#### Model Selection

```python
from sklearn.feature_selection import SelectFromModel
```

```python
import warnings
warnings.filterwarnings('ignore')

# Fit model using each importance as a threshold
thresholds = sorted(clf.feature_importances_)
for thresh in thresholds:
    # select features using threshold
    selection = SelectFromModel(clf, threshold=thresh, prefit=True)
    select_X_train = selection.transform(X_train)
    # train model
    selection_model = XGBClassifier()
    selection_model.fit(select_X_train, y_train)
    # eval model
    select_X_val = selection.transform(X_val)
    y_pred = selection_model.predict(select_X_val)
    y_prob = selection_model.predict_proba(select_X_val)
    accuracy = accuracy_score(y_val, y_pred)
    loss = log_loss(y_val, y_prob)
    print("Thresh=%.3f, n=%d, Accuracy: %.2f%%, Loss: %.3f" % (thresh, select_X_train.shape[1], accuracy*100.0, loss))
```

#### Evaluate Model

```python
thresh = 0.031

# select features using threshold
selection = SelectFromModel(clf, threshold=thresh, prefit=True)
select_X_train = selection.transform(X_train)
# train model
selection_model = XGBClassifier()
selection_model.fit(select_X_train, y_train)
# eval model
select_X_test = selection.transform(X_test)
y_pred = selection_model.predict(select_X_test)
y_prob = selection_model.predict_proba(select_X_test)
accuracy = accuracy_score(y_test, y_pred)
loss = log_loss(y_test, y_prob)
print("Thresh=%.3f, n=%d, Accuracy: %.2f%%, Loss: %.3f" % (thresh, select_X_train.shape[1], accuracy*100.0, loss))
```
